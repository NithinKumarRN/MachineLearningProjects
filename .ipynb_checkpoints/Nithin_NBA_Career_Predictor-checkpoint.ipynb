```python
## Defining function for checking correlations further

def correlation(df,level=0.85):
    corr_df = df.corr()
    for i in corr_df.columns:
        if any( (corr_df[i]>level).sort_values(ascending=False)[1:] ):
            print(i)
            print(corr_df[corr_df[i]>level][i].sort_values(ascending=False)[1:])
        print('-'*50)
```

```python
from scipy.stats import shapiro,normaltest,anderson

def normality_tests(df,col):
    """Returns the results of Normality tests for Shapiro Wilk,D'Agostino,Anderson
       and seaborn distribution plot of the variable"""

    print("Shapiro: ",shapiro(df[col].dropna().sample(40,random_state=0)))
    print()
    print(normaltest(df[col].dropna().sample(40,random_state=0)))
    print()
    print(anderson(df[col].dropna().sample(40,random_state=0)))
    sns.distplot(df[col].dropna())
    plt.show()
```

```python
normality_tests(nba_df,'3P%')  ## All the test indicate to a non parametric distribution
```

```python
X.drop(['FG%','3P%','FT%','REB'],axis=1,inplace=True)
```

```python
## Checking distribution of variables for patterns, abnormality etc.

for i in X.columns:
    sns.distplot(X[i],)
    plt.show()
```

```python
## Scaling data to standarize.

X_scaled = StandardScaler().fit_transform(X.values.astype('float64'))

X_scaled = pd.DataFrame(X_scaled,columns=X.columns)
```

```python
def cross_validation(estimator,X,y,metric=None):
    
    """Returns model.score value for the Estimator specified"""

    scores = cross_val_score(estimator,X,y,cv=10)
    #print( "\nAccuracy scores\n\n", scores )
    
    #return scores.mean()

    print( "\nMean Accuracy", scores.mean() )    
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled,y, test_size=0.33, random_state=42)
```

```python
## Here we are only taking baseline models of each algo and not attempting a Gridsearch due to computing limitations.

models = [LogisticRegression(solver='liblinear',multi_class='ovr'),RidgeClassifier(),
          LinearDiscriminantAnalysis(),KNeighborsClassifier(),DecisionTreeClassifier(),
          RandomForestClassifier(n_estimators=300),AdaBoostClassifier(n_estimators=300),
          GradientBoostingClassifier(),SVC(gamma='auto'),XGBClassifier(n_estimators=300)]

model_names = ['LogisticRegression','RidgeClassifier','LinearDiscriminantAnalysis','KNeighborsClassifier',
               'DecisionTreeClassifier','RandomForestClassifier','AdaBoostClassifier',
               'GradientBoostingClassifier','SVC','XGBClassifier']

combine = list(zip(model_names,models))
```

```python
for i,v in combine:
    print(i)
    cross_validation(v,X_train,y_train)
    print()
    print('-'*50)
```

```python
pred_dataframe = pd.DataFrame()
for model_name,model in combine:
    model.fit(X_train,y_train)
    pred = model.predict(X_test)
    pred_dataframe[model_name] = pred
```

```python
pred_dataframe['actual'] = y_test.values
pred_dataframe.head()
```

```python
## Trying wisdom of the crowd method by taking mode of the predictions of different classifiers

mode_lst = []
for i in range(pred_dataframe.shape[0]):
    mode_lst.append(pred_dataframe.iloc[i,:].mode()[0])

pred_dataframe['mode']= mode_lst

accuracy_score(pred_dataframe['actual'],pred_dataframe['mode'])
```

```python
## By just combining the outputs of all the models and taking the majority votes we acheived an increase in accuracy by 4%.
```

**ENSEMBLE LEARNING : Stacking Models**

```python
from sklearn.model_selection import StratifiedKFold ## Importing StratifiedKFold Library for random sample splits.
```

```python
### Defining a stacking function which takes in the estimator, Xtrain and y train, splits into folds and validates 
### the prdictions on validation fold, rturns th predictions, accuracy of validation set prediction and Y values involved.
### Also returns the model's prediction of the Xtest.


def stacking(model,X_train,y_train,X_test,n_folds):
    
    folds = StratifiedKFold(n_splits=3,random_state=2) ## n_splits can be increased if enough data is involved
    
    test_pred = np.empty((0,1))
    
    train_pred = np.empty((0,1))
    
    for train_indices, test_indices in folds.split(X_train,y_train.values):
        
        x_train , x_val = X_train.iloc[train_indices] , X_train.iloc[test_indices]
        
        Y_train , Y_val = y_train.iloc[train_indices] , y_train.iloc[test_indices]
    
    model.fit(x_train,Y_train)
    
    train_pred = np.append( train_pred,model.predict(x_val) )
    
    test_pred = np.append( test_pred,model.predict(X_test) )
    
    score = accuracy_score(train_pred.astype(int),Y_val.astype(int))
    
    return train_pred , test_pred , Y_train , Y_val , score
```

```python
### Output function returns a dataframe of baseline (Level0) models' predictions on Validation Set (for fitting Level1 Ensemble)
### and Test sets to create a dataframe to be used as X_test for Level1 Estimator
### Note that models are present in combine which is a zipped form of (modelnames,models) ) .
### Y_val is the label for validation set dataframe and is needed for training Level1 Estimator
### If scoring is enabled, prints validation set scores of baseline models.

def output(t,scoring=False):

    d = {}

    for i,v in combine:


        train_pred , test_pred , Y_train , Y_val , score = stacking(v,X_train,y_train,X_test,n_folds=5)
        
        if t=='train':
            
            d[i] = train_pred
            
        elif t=='test':
            
            d[i] = test_pred
            
        if scoring:
            print(i , ': Validation set score')
            print()
            print(score)
            print('-'*30)

    return pd.DataFrame.from_dict(d), Y_val
```

```python
## Validation set results of different baseline classifiers.

df,y_df =  output('train',scoring=True)
```

```python
## Create a List to chose the best or most diverse baseline models to train the Level Stack Model.

cols  = ['LogisticRegression','SVC','GradientBoostingClassifier','DecisionTreeClassifier']
```

```python
## Training XGBoost model with the pick of the baseline models

xgb = XGBClassifier()
xgb.fit(df[cols],y_df)

## Creating X_test dataframe consisting of baseline models' predictions of original X_test. 
## Level 1 XGBoost model score for the Test set.

df_test , _ = output('test')
xgb.score(df_test[cols],y_test)
```
